```python
from pycaret.datasets import get_data
```

```python
dataset=get_data('diamond')
```

```python
dataset.shape
```

```python
dataset.columns
```

```python
dataset.describe()
```

# spilting data for train

```python
data_train = dataset.sample(frac= 0.9, random_state=786)
```

```python

data_train.shape
```

# spilting data for test

```python
data_test = dataset.drop(data_train. index)
```

```python

data_test.shape
```

# arranging dataset accordingly for data_train

```python
data_train.reset_index(inplace=True,drop=True)
```

# arranging dataset accordingly for data_test

```python
data_test.reset_index(inplace=True,drop=True)
```

# implemenation of linear regression

```python
from pycaret.regression import *
```

```python

reg = setup(data= data_train,
target = "Price",
session_id=123)
```

```python
best=compare_models(exclude=['ransac'])
```

```python
gbr=create_model('gbr')
```

```python
tuned_gbr=tune_model(gbr)
print(tuned_gbr)
```

# plotting the r2 score

```python
plot_model(tuned_gbr)
```

# plotting the error for Gradient Boosting Regressor

```python
plot_model(tuned_gbr, plot="error")
```

# plotting all the important features for regression model

```python
plot_model(tuned_gbr, plot="feature")
```

# making prediction validation data

```python
predict_model(tuned_gbr)
```
